## Playlist Name Analysis

```python
################# Importation des playlists utiles  #################

def names(path, num_slices=20):
  cnt = 0 # We initiate the number of file to 0
  mylist = []
  filenames = os.listdir(path) # get all the file paths

  for fname in sorted(filenames):
    print(fname)

    cnt += 1
    fullpath = os.sep.join((path, fname))
    f = open(fullpath)
    js = f.read()
    f.close()
    current_slice = json.loads(js)      
      
    # Create a list of all playlists
    for playlist in current_slice['playlists']:
      mylist.append(playlist["name"])
    if cnt == num_slices:
      break

  return mylist # Path where the json files are extracted
```

```python
def loop_slices(path, category, num_slices=20):
  cnt = 0 # We initiate the number of file to 0
  mpd_playlists = pd.DataFrame(columns=["playlist_id", "playlist_name", "track_number", "track_uri"]) # Create the DataFrame for the playlists
  filenames = os.listdir(path) # get all the file paths
  id = 0 

  for fname in sorted(filenames):
    print(cnt)
    cnt += 1

    fullpath = os.sep.join((path, fname))
    f = open(fullpath)
    js = f.read()
    f.close()
    current_slice = json.loads(js)      
      
    # Create a list of all playlists
    for playlist in current_slice['playlists']:
      if len([e for e in category if e in playlist["name"]]) > 0:
        for track in playlist["tracks"]:
          entry = pd.DataFrame({"playlist_id": id, "playlist_name": playlist["name"], "track_number": track["pos"], "track_uri": track['track_uri']},index=[0])
          mpd_playlists = pd.concat([mpd_playlists, entry])
        id = id + 1
    
    if cnt == num_slices:
      break

  return mpd_playlists# Path where the json files are extracted
```
